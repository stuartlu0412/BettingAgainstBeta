```python
import numpy as np
import pandas as pd
import yfinance as yf
import quantstats as qs
import matplotlib.pyplot as plt

from data_loading import *

pd.options.mode.copy_on_write = True
```

```python
# 1. 解決太長的欄位
# 2. 將年月日轉換成 DateTime
# 3. 將 (證券代碼, Date) 設定為 index
# 4. 將所有的欄位轉換成數值型態
# 5. 存至 sqlite
```

```python

# List of file paths for the 20 CSV files
file_paths = ['./data/raw/2000-2002.csv',
              './data/raw/2002-2004.csv',
              './data/raw/2004-2006.csv',
              './data/raw/2006-2008.csv',
              './data/raw/2008-2010.csv',
              './data/raw/2010-2015.csv',
              './data/raw/2015-2018.csv',
              './data/raw/2018-2020.csv',
              './data/raw/2020-2021.csv', 
              './data/raw/2021-2022.csv',
              './data/raw/2022-2023.csv',
              './data/raw/2023-2024.csv',
              './data/raw/2024-2025.csv',          
              ]  # Add all 20 file paths

# Initialize an empty list to store data frames
dfs = []

# Load and process each data frame
for file_path in file_paths:
    df = pd.read_csv(file_path, encoding='cp950', usecols=range(0, 36))
    df['年月日'] = pd.to_datetime(df['年月日'], format='%Y%m%d')
    df.rename(columns={'證券代碼': '證券代碼', '年月日': 'Date', '收盤價(元)': 'Close'}, inplace=True)
    df.set_index(['證券代碼', 'Date'], inplace=True)
    df.sort_index(inplace=True)
    dfs.append(df)
```

```python
# Combine all data frames into a single data frame
combined_df = pd.concat(dfs, axis=0) 
combined_df
```

```python
otc_stocks = combined_df[combined_df['上市別'] == 'OTC']

pd.to_pickle(otc_stocks, './data/processed/otc_stocks.pkl')
```

```python
tse_stocks = combined_df[combined_df['上市別'] == 'TSE']
pd.to_pickle(tse_stocks, './data/processed/tse_stocks.pkl')
```
